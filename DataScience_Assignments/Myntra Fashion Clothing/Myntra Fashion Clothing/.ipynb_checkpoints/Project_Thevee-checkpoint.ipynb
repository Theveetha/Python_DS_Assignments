```python
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
color_group = color_group.sort_values(by = "totalRecords", ascending = False)
sns.barplot(x=color_group.index.values,y=color_group["totalRecords"].values,palette="magma")
plt.xticks(rotation=90,ha="right")
plt.title("Color Vs totalRecords")


plt.subplot(1,3,2)
color_group = color_group.sort_values(by = "AvgPrice", ascending = False)
sns.barplot(x=color_group.index.values,y=color_group["AvgPrice"].values,palette="magma")
plt.xticks(rotation=90,ha="right")
plt.title("Color Vs AvgPrice")


plt.subplot(1,3,3)
color_group = color_group.sort_values(by = "TotalPrice", ascending = False)
sns.barplot(x=color_group.index.values,y=color_group["TotalPrice"].values,palette="magma")
plt.xticks(rotation=90,ha="right")
plt.title("Color Vs TotalPrice")
plt.show()
```

```python
# Using for loop
col_Name = color_group.index.values
# print(col_Name)
pos = color_group.columns.values
# print(pos)

plt.figure(figsize=(15,3))
for i in range(len(pos)):
    plt.subplot(1,3,i+1)
    color_group = color_group.sort_values(by = pos[i], ascending = False)
    col_Name = color_group.index.values
    for j in range(len(col_Name)):
        sns.barplot(x=col_Name,y=color_group[pos[i]].values,palette="magma")
        plt.xticks(rotation=90,ha="right")
        plt.title(f"Color vs {pos[i]}")
plt.show()
```

```python
# Using for loop but picking only top 10 records
col_Name = color_group.index.values[:10]
# print(col_Name)
pos = color_group.columns.values
# print(pos)

plt.figure(figsize=(15,3))
for i in range(len(pos)):
    plt.subplot(1,3,i+1)
    color_group = color_group.sort_values(by = pos[i], ascending = False)
    col_Name = color_group.index.values[:10]
    for j in range(len(col_Name)):
        sns.barplot(x=col_Name,y=color_group[pos[i]].values[:10],palette="magma")
        plt.xticks(rotation=30,ha="right")
        plt.title(f"Color vs {pos[i]}")
plt.show()
```

```python
# Here we are just seeing only top 10
col_Name = color_group.index.values[:6]
# print(col_Name)
pos = color_group.columns.values
# print(pos)
# print(colors)
plt.figure(figsize=(15,3))
for i in range(len(pos)):
    plt.subplot(1,3,i+1)
    color_group = color_group.sort_values(by = pos[i], ascending = False)
    col_Name = color_group.index.values[:6]
    colors = list(map(lambda x : "cadetblue" if x == "Others" else x,col_Name))
    colors = list(map(lambda x : "pink" if x == "Rose" else x,colors))
    colors = list(map(lambda x : "dimgrey" if x == "Platinum" else x,colors))
#     print(colors)
# The above mentioned we can use to give the color palatte of our choice 
# since few colors name vary from predefined name we are changing it multiple times
    for j in range(len(col_Name)):
        plt.pie(x=color_group[pos[i]].values[:6],labels=col_Name,startangle=0
                ,counterclock=False,colors=colors)
#         plt.xticks(rotation=90,ha="right")
        plt.title(f"Color vs {pos[i]}")
plt.show()
```
